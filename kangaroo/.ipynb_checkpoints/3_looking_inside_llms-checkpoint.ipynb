```python
#234567890#234567890#234567890#234567890#234567890#234567890#234567890#234567890
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
mod = 'microsoft/Phi-3-mini-4k-instruct'
tokenizer = AutoTokenizer.from_pretrained(mod)
model = AutoModelForCausalLM.from_pretrained(
    mod, device_map='auto', torch_dtyep='auto', trust_remote_code=True)
generator = pipeline(
    'text-generation', 
    model=model, 
    tokenizer=tokenizer 
    return_full_text=False,
    max_new_tokens=50,
    do_sample=False)
```

```python
prompt = (
    'Write an email apologizing to Sarah for the tragic gardening '
    'mishap. Explain how it happened.')
output = generator(prompt)
output[0]['generated_text']
```
